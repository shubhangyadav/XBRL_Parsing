```python
from bs4 import BeautifulSoup
from collections import defaultdict

tags=[]
html = open("Desktop/HTML Files/Prod223_2706_00088430_20191231.html").read()
soup = BeautifulSoup(html, 'lxml')
tag_list = soup.find_all('ix:nonfraction')
```

```python
for i in tag_list:
    try:
        tags.append(str(i).split()[4].split(':')[1].split('"')[0])
        tags.append(i.text)
        tags.append(str(i).split()[1])
    except IndexError:
        tags.append(str(i).split()[3].split(':')[1].split('"')[0])
        tags.append(i.text)
        tags.append(str(i).split()[1])
tags
```

```python
def separate(lst, n):
    return zip(*[iter(lst)]*n)
```

```python
separated_list=[]
for values in separate(tags, 3):
    print(list(values))
    separated_list.append(list(values))
```

```python
d = defaultdict(list)

for i, j, k in separated_list:
    d[i].append(j)
    d[i].append(k)

merged_by_tag = [[k]+v for k, v in d.items()]
merged_by_tag
```

```python
def convert(lst):
    lst1=[lst[i] for i in range(1, len(lst),1)]
    res_dct= {lst[0]: [lst1[x:x+4] for x in range(0, len(lst1),4)]}
    return res_dct
```

```python
final=[]
for lst in merged_by_tag:        
    #print(convert(lst))
    final.append(convert(lst))
final
```
